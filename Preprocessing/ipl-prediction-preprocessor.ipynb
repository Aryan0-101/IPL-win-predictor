city Nan = Dubai**


```python
mat_df['city'] = mat_df['city'].fillna('Dubai')
```

# remove no result


```python
mat_df = mat_df.drop(mat_df[mat_df['winner'].isnull()].index)
```

# D/L affected removed 

```python
mat_df = mat_df[mat_df['method'].isnull()]
```

```python
mat_df['result_margin'] = mat_df['result_margin'].fillna(0)
```

# Getting only the current teams

```python
mat_df['team1'].unique()
```

```python
teams = ['Royal Challengers Bengaluru',
        'Punjab Kings',
        'Delhi Capitals',
        'Lucknow Super Giants',
        'Sunrisers Hyderabad',
        'Mumbai Indians',
        'Gujarat Titans',
        'Chennai Super Kings',
        'Rajasthan Royals',
        'Kolkata Knight Riders']
```

```python
del_df['batting_team'] = del_df['batting_team'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')
del_df['bowling_team'] = del_df['bowling_team'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')

del_df['batting_team'] = del_df['batting_team'].str.replace('Kings XI Punjab', 'Punjab Kings')
del_df['bowling_team'] = del_df['bowling_team'].str.replace('Kings XI Punjab', 'Punjab Kings')

del_df['batting_team'] = del_df['batting_team'].str.replace('Delhi Daredevils', 'Delhi Capitals')
del_df['bowling_team'] = del_df['bowling_team'].str.replace('Delhi Daredevils', 'Delhi Capitals')

del_df['batting_team'] = del_df['batting_team'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
del_df['bowling_team'] = del_df['bowling_team'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

del_df['batting_team'] = del_df['batting_team'].str.replace('Gujarat Lions', 'Gujarat Titans')
del_df['bowling_team'] = del_df['bowling_team'].str.replace('Gujarat Lions', 'Gujarat Titans')
```

```python
del_df = del_df[del_df['batting_team'].isin(teams)]
del_df = del_df[del_df['bowling_team'].isin(teams)]
```

```python
mat_df['match_id'] = mat_df['id']
```

```python
mat_df = mat_df.drop('id', axis = 1)
mat_df['match_id'].astype(int)
```

**Merging the Matches data with the ball-by-ball data to get the city,winner and target runs for each match with each delivery.**

```python
mat_df = mat_df[['match_id','city', 'winner', 'target_runs']]
delivery_df = mat_df.merge(del_df, on = 'match_id')
```

```python
delivery_df['inning'] = pd.to_numeric(delivery_df['inning'], errors='coerce')
```

**Taking only the required variables from the data**

```python
delv_df = delivery_df.drop(["batter","bowler", "non_striker", "batsman_runs" , "dismissal_kind", "fielder", "player_dismissed"], axis = 1)
delv_df['total_runs'] = pd.to_numeric(delv_df['total_runs'], errors='coerce')
delv_df = delv_df[delv_df["inning"] == 2]
delv_df.info()
```

# Adding column for the current score

```python
delv_df['current_score']=delv_df.groupby('match_id')['total_runs'].cumsum()
```

# # **Adding column for the Runs left**  

```python
delv_df['runs_left'] = delv_df['target_runs'] - delv_df['current_score']
```

# # **Adding column for balls left**

```python
delv_df['balls_left'] = 120 - (delv_df['over'] * 6 + delv_df['ball'])
```

# **Adding column for wickets left**

```python
delv_df['is_wicket'] = delv_df['is_wicket'].astype(int)
wickets = delv_df.groupby('match_id')['is_wicket'].cumsum()
delv_df['wickets_left'] = 10 - wickets
```

# # Adding column for Current Run Rate 

```python
delv_df['RR'] = (delv_df['current_score']*6) / (120 - delv_df['balls_left'])
```

# # Adding column for Required Run Rate 

```python
delv_df['RRR'] = (delv_df['runs_left'] *6 )/ delv_df['balls_left']
```

# Adding column for the result i.e if the chasing team won the match or not. 

```python
def result(row): 
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delv_df['result'] = delv_df.apply(result, axis = 1)
```

```python
delv_df
```

# **The final Data**

```python
data = delv_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets_left', 'target_runs', 'RR', 'RRR', 'result']]
```

```python
data['city'] = data['city'].replace('Bangalore', 'Bengaluru')
data['city'] = data['city'].replace('Navi Mumbai', 'Mumbai')
```

```python
data['city'].unique()
```

```python
data = data.sample(data.shape[0])
```

```python
data.to_csv("ipl_data.csv", index =False)
```
